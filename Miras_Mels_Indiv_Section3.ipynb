```python
cars = ['Mercedes', 'Ford', 'Mazda', 'BMW']
print(cars)
```

```python
cars.append('Honda')
print(cars)
```

```python
ford_index = cars.index('Ford')
cars[ford_index] = 'Toyota'
print(cars)
```

```python
inventory = {
    "Product": ["Laptop", "Monitor", "Keyboard", "Mouse"],
    "Quantity": [150, 400, 750, 600],
    "Price": [1200.50, 350.75, 75.00, 25.50],
    "Available": [True, True, False, True]
}
print(inventory)
```

```python
num_records = len(inventory["Product"])
second_product = inventory["Product"][1]
first_price = inventory["Price"][0]
third_availability = inventory["Available"][2]
print(f"\nThere are {num_records} records in the dictionary.")
print(f"The second product is {second_product}, the price of the first product is {first_price}, and it is {third_availability} that the third product is available.")
```

```python
print("\nData Types of Dictionary Variables:")
for key, value in inventory.items():
    print(f"The data type of '{key}' is: {type(value)}")
```
